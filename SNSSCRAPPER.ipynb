```python
import csv
import snscrape.modules.twitter as sntwitter
import pandas as pd
import itertools
```

```python
def ultimatescrapper(text = '#aps', start = '2016-1-1', end = '2022-1-16', radius = '5000km', numberOfTweets = 2000):
    scraped_tweets = sntwitter.TwitterSearchScraper(f'{text} since:{start} until:{end} near:"pakistan"').get_items()
    sliced_scraped_tweets = itertools.islice(scraped_tweets, numberOfTweets)
    tweets_df = pd.DataFrame(sliced_scraped_tweets)
    tweets_all_df = pd.concat([ tweets_df], axis=0)
    tweets_all_df.to_csv("aps.csv", index=False)
    return tweets_df
```

```python
print(ultimatescrapper())
```
